```python
import pandas as pd

# 读取数据
df = pd.read_csv('E-commerce data.csv', encoding='ISO-8859-1')  # 注意这个数据集默认不是utf-8

# 查看前5行
df.head()
```

```python
# 查看数据类型和缺失值
print("数据总览：")
print(df.info())
print("\n缺失值统计：")
print(df.isnull().sum())

# 删除缺失 CustomerID 的记录（无法追踪用户行为）
df = df.dropna(subset=['CustomerID'])

# 去除 Quantity 为负的记录（退货单）
df = df[df['Quantity'] > 0]

# 去除价格为 0 或负的记录
df = df[df['UnitPrice'] > 0]

# 转换日期格式
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 将 CustomerID 转为字符串类型
df['CustomerID'] = df['CustomerID'].astype(str)

# 保存清洗后的数据
df.to_csv('cleaned_data.csv', index=False)
print("✅ 数据清洗完成，已保存为 cleaned_data.csv")
```
